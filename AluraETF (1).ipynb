##          Relatório Final – Projeto Cancelamento de Clientes da Telecom X

## Introdução
EA Telecom X enfrenta desafios relacionados à retenção de clientes. Este projeto tem como objetivo realizar uma Análise Exploratória de Dados (EDA) nos dados de clientes fornecidos para identificar padrões, tendências e fatores que contribuem para o "churn" (evasão de clientes). A partir dessa análise, espera-se gerar insights acionáveis que possam embasar estratégias de retenção mais eficazes.

## Limpeza e Tratamento de Dados

🧹 2. Limpeza e Tratamento de Dados

Os dados foram inicialmente obtidos a partir de uma URL no formato JSON. As principais etapas de preparação e limpeza incluíram:

Normalização: A estrutura JSON foi normalizada em um DataFrame tabular. O DataFrame inicial continha 7267 linhas e 21 colunas.

Conversão de Tipos de Dados:
A coluna account.Charges.Total foi identificada como object (texto) e continha valores vazios (''). Esses valores foram substituídos por np.nan e a coluna foi convertida para o tipo float64.

Colunas binárias (Yes/No) foram convertidas para 1 e 0 (inteiros).

A variável alvo Churn foi convertida para Int8, com Yes como 1, No como 0, e valores vazios como np.nan.

Tratamento de Valores Ausentes: Foi verificada a presença de 224 valores nulos na coluna Churn. Outras colunas também apresentavam valores nulos, incluindo account.Charges.Total (11 nulos). Foram identificadas 235 linhas com valores nulos.

As linhas com valores nulos foram removidas do DataFrame, resultando em 7032 linhas e 22 colunas.


Criação de Novas Variáveis: Uma nova coluna chamada Contas_Diarias foi criada, calculada a partir de account.Charges.Monthly dividido por 36.

Renomeação de Colunas: As colunas foram renomeadas para facilitar a compreensão e o uso, por exemplo, 'customerID' para 'id_cliente', 'Churn' para 'cancelou_servico', 'customer.gender' para 'genero', entre outras.

Padronização de Categorias: Os valores na coluna genero foram traduzidos de 'Female' para 'Feminino' e 'Male' para 'Masculino'. Os metodo_pagamento foram traduzidos para português. Os tipo_contrato foram traduzidos para 'Mensal', 'Anual' e 'Bienal'.

As categorias de serviços como 'No internet service' ou 'No phone service' foram padronizadas para 'No' e, posteriormente, convertidas para 0, enquanto 'Yes' foi convertido para 1.

## 3. Análise Exploratória de Dados (EDA)

A análise exploratória focou na variável cancelou_servico (Churn) e na relação com outras características dos clientes.

📌 **3.1. Perfil dos Clientes que Evadem**



*   Distribuição Geral do Churn: Aproximadamente 26.6%
dos clientes cancelaram o serviço, enquanto 73.4% permaneceram ativos.
*   Gênero: A taxa de cancelamento é bastante similar entre gêneros, com 27.0% para o gênero Feminino e 26.2% para o Masculino, indicando que o gênero não é um fator de grande distinção para o churn.

*   Idosos e Dependentes:
         Clientes idosos (idoso=1) apresentam uma taxa de cancelamento significativamente maior em comparação com os não-idosos (idoso=0).

         Clientes sem dependentes (tem_dependentes=0) tendem a cancelar mais do que aqueles com dependentes (tem_dependentes=1).


*   Serviço de Internet: Clientes com serviço de internet "Fiber optic" (fibra óptica) apresentam a maior taxa de cancelamento (61.3%), seguidos por "DSL" (27.8%). Clientes sem serviço de internet ('0') têm a menor taxa de churn (10.9%).

*   Tipo de Contrato: Clientes com contrato "Mensal" mostram uma taxa de cancelamento muito mais elevada em comparação com contratos "Anual" e "Bienal".
Método de Pagamento: O "Cheque eletrônico" é o método de pagamento com a maior taxa de churn.*

## Conclusões e Insights


*   A taxa de churn na Telecom X é relevante, afetando cerca de um quarto da base de clientes.
*   Idosos e clientes sem dependentes são mais propensos ao cancelamento.


*   O tipo de serviço de internet "Fiber optic" e o contrato "Mensal" são fortes indicadores de alta propensão ao churn
*   O método de pagamento "Cheque eletrônico" está associado a uma maior taxa de evasão.

* A "fidelidade" do cliente, medida pelo meses_contrato (tenure), mostra que clientes com menor tempo de contrato tendem a ter maior taxa de churn (implícito pelas análises de meses_contrato e faturamento_total)







## Recomendações

* Programas de Fidelidade e Retenção para Idosos e Clientes Sem Dependentes: Desenvolver campanhas direcionadas para esses segmentos, oferecendo benefícios ou suporte diferenciado.

* Estratégias para Contratos "Fiber Optic" e "Mensal": Investigar os motivos do alto churn nesses grupos. Para contratos mensais, incentivar a migração para contratos de longo prazo (anual ou bienal) com descontos ou vantagens. Para "Fiber optic", pode ser necessário avaliar a qualidade do serviço ou a competitividade dos preços.

* Análise do Método de Pagamento "Cheque Eletrônico": Compreender por que esse método de pagamento está associado a maior churn. Pode haver problemas de usabilidade, taxas ou outros fatores que desestimulem a permanência.

* Monitoramento da Jornada do Cliente: Clientes com menor tempo de contrato (menor meses_contrato) necessitam de atenção especial nos primeiros meses para garantir a satisfação e reduzir a probabilidade de churn inicial.
